# Objetivo do Trabalho
O objetivo deste trabalho √© explorar a aplica√ß√£o de t√©cnicas de Machine Learning para identificar tend√™ncias de mercado em s√©ries temporais financeiras altamente ruidosas, utilizando a√ß√µes representativas do S&P 500. A abordagem trata os pre√ßos das a√ß√µes como sinais financeiros compostos por uma componente de tend√™ncia estrutural e uma componente de ru√≠do, refletindo a incerteza do mercado e o comportamento agregado dos participantes.


## O trabalho ambiciona:

- Demonstrar a viabilidade de extrair padr√µes estat√≠sticos √∫teis em dados financeiros ruidosos, mesmo quando n√£o √© poss√≠vel prever movimentos de pre√ßos pontuais.

- Treinar um modelo de Machine Learning que combine informa√ß√µes sobre tend√™ncia e ru√≠do para estimar a probabilidade de movimentos direcionalmente positivos ou negativos do mercado.

- Aplicar o modelo a dados atuais para inferir tend√™ncias agregadas do mercado, apresentando uma vis√£o interpret√°vel (ex.: tend√™ncia de subida, descida ou neutra).

- Discutir os limites e implica√ß√µes desta abordagem, incluindo a influ√™ncia do ru√≠do, volatilidade e regime de mercado, enfatizando que o objetivo n√£o √© gerar lucro ou previs√£o exata, mas sim uma prova de conceito acad√©mica robusta.

### Outros Objetivos

- Dire√ß√£o do Mercado (Classifica√ß√£o Bin√°ria/Multiclasse)
>Sobe ou Desce no pr√≥ximo dia (classifica√ß√£o bin√°ria)
Alta, Baixa ou Lateral (classifica√ß√£o multiclasse)
Exemplo: "Amanh√£ a AAPL vai subir ou cair?"
>
- Volatilidade
>Prever se o pr√≥ximo dia ser√° vol√°til ou est√°vel
>
- Faixas de Pre√ßo (Regress√£o por Intervalos)
>Prever em qual faixa o pre√ßo vai estar (ex: $220-225, $225-230)

# Expectativa de Resultados

## 1. Hip√≥tese do Mercado Eficiente (EMH)

Segundo a teoria de **Eugene Fama**:

- **Mercados eficientes:** pre√ßos refletem **toda informa√ß√£o dispon√≠vel**.
- **Limite te√≥rico de acur√°cia:** ~50% (equivalente ao acaso puro).
- **Realidade:** Mercados s√£o geralmente **semi-eficientes**.

---

## 2. Consenso Acad√™mico e Profissional

| Horizonte       | Acur√°cia Realista | Acur√°cia "Excelente" | Observa√ß√µes |
|-----------------|-----------------|--------------------|------------|
| **Intraday (minutos)** | 52-55% | 58-62% | Alta frequ√™ncia, muito ru√≠do |
| **Di√°rio (D+1)** | 52-56% | 58-65% | Padr√µes t√©cnicos fracos |
| **Semanal (D+5)** | 55-60% | 65-70% | Tend√™ncias mais claras |
| **Mensal (D+20)** | 60-65% | 70-75% | Fundamentos dominam |

> Usando **apenas pre√ßos hist√≥ricos (OHLCV)**, o **limite pr√°tico** de acur√°cia √© **55-60%**.

### Limita√ß√µes de Dados
- **Fundamentos:** balan√ßos, lucros, guidance
- **Sentimento:** not√≠cias, redes sociais, earnings calls
- **Orderflow:** fluxo de ordens, market microstructure
- **Macroecon√¥micos:** Fed, infla√ß√£o, juros

---

## 3. Horizonte de Predi√ß√£o

- **Di√°rio (D+1):** mais dif√≠cil, muito ru√≠do  
- **Semanal (D+5):** +5-10% de acur√°cia potencial  
- **Mensal (D+20):** +10-15% de acur√°cia potencial  

---

## 4. Modelo de Decomposi√ß√£o de Sinais Financeiros

\[
\text{Pre√ßo}(t) = \text{Tend√™ncia}(t) + \text{Sentimento}(t) + \text{Ru√≠do}(t)
\]

| Componente       | Descri√ß√£o |
|-----------------|------------|
| **Tend√™ncia(t)** | Estrutura de longo prazo (previs√≠vel) |
| **Sentimento(t)** | Comportamento agregado dos participantes (semi-previs√≠vel) |
| **Ru√≠do(t)**      | Aleatoriedade irredut√≠vel (imprevis√≠vel) |

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# ML
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (accuracy_score, balanced_accuracy_score, f1_score, 
                             roc_auc_score, confusion_matrix, classification_report)
from sklearn.model_selection import TimeSeriesSplit
from sklearn.utils import resample

# Estat√≠stica
from scipy.stats import permutation_test, chi2_contingency
from scipy import stats

# Dados
import yfinance as yf

# ML avan√ßado
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

print("="*80)
print("PIPELINE DE ML PARA MERCADOS FINANCEIROS - VERS√ÉO ACAD√âMICA ROBUSTA")
print("="*80)
```

**Vari√°veis Globais**

```python
# Lista de a√ß√µes do S&P 500 - Empresas de tecnologia
tech_list = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META', 'NVDA', 'TSLA', 'NFLX']

end = datetime.now()
start = end - timedelta(days = 5 * 365) # 5 anos
company_data = {}
```

```python
# Baixar os dados das a√ß√µes
company_data = {}
for stock in tech_list:
    try:
        print(f"  Downloading {stock}...", end=" ")
        df = yf.download(stock, start=start, end=end, progress=False)
        if len(df) > 0:
            if isinstance(df.columns, pd.MultiIndex):
                df.columns = df.columns.get_level_values(0)
            df = df.reset_index()
            df['ticker'] = stock
            company_data[stock] = df
            print(f"{len(df)} dias ‚úì")
        else:
            print("Sem dados !")
    except Exception as e:
        print(f"Erro: {e}")

df_raw = pd.concat(company_data.values(), axis=0, ignore_index=True)
print(f"\nTotal: {len(df_raw)} registros")
print(f"  Per√≠odo: {df_raw['Date'].min().date()} at√© {df_raw['Date'].max().date()}")
```

### Decomposi√ß√£o: Tend√™ncia e Ruido

**Visualiza√ß√µes** 

```python
print("\n[2/10] FEATURE ENGINEERING (VERIFICA√á√ÉO DE DATA LEAKAGE)")
print("-"*80)

def create_features_no_leakage(df):
    """
    Cria features garantindo que n√£o h√° data leakage.
    REGRA: Apenas informa√ß√£o dispon√≠vel AT√â t-1 pode ser usada para prever t
    """
    df = df.copy()
    df = df.sort_values(['ticker', 'Date']).reset_index(drop=True)
    
    # Retornos (shift para garantir que usamos pre√ßo ANTERIOR)
    df['Return_1d'] = df.groupby('ticker')['Close'].pct_change(1)
    df['Return_5d'] = df.groupby('ticker')['Close'].pct_change(5)
    df['Return_10d'] = df.groupby('ticker')['Close'].pct_change(10)
    
    # M√©dias M√≥veis (rolling j√° olha para tr√°s por padr√£o)
    df['SMA_5'] = df.groupby('ticker')['Close'].transform(lambda x: x.rolling(5, min_periods=5).mean())
    df['SMA_10'] = df.groupby('ticker')['Close'].transform(lambda x: x.rolling(10, min_periods=10).mean())
    df['SMA_20'] = df.groupby('ticker')['Close'].transform(lambda x: x.rolling(20, min_periods=20).mean())
    df['SMA_50'] = df.groupby('ticker')['Close'].transform(lambda x: x.rolling(50, min_periods=50).mean())
    
    # Cruzamentos (features derivadas)
    df['SMA_Cross_5_20'] = (df['SMA_5'] - df['SMA_20']) / df['SMA_20']
    df['SMA_Cross_10_50'] = (df['SMA_10'] - df['SMA_50']) / df['SMA_50']
    
    # RSI
    def calculate_rsi(prices, period=14):
        delta = prices.diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=period, min_periods=period).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=period, min_periods=period).mean()
        rs = gain / loss
        rsi = 100 - (100 / (1 + rs))
        return rsi
    
    df['RSI_14'] = df.groupby('ticker')['Close'].transform(lambda x: calculate_rsi(x, 14))
    
    # Bollinger Bands
    df['BB_Middle'] = df.groupby('ticker')['Close'].transform(lambda x: x.rolling(20, min_periods=20).mean())
    df['BB_Std'] = df.groupby('ticker')['Close'].transform(lambda x: x.rolling(20, min_periods=20).std())
    df['BB_Upper'] = df['BB_Middle'] + (df['BB_Std'] * 2)
    df['BB_Lower'] = df['BB_Middle'] - (df['BB_Std'] * 2)
    df['BB_Position'] = (df['Close'] - df['BB_Lower']) / (df['BB_Upper'] - df['BB_Lower'] + 1e-10)
    
    # MACD
    df['EMA_12'] = df.groupby('ticker')['Close'].transform(lambda x: x.ewm(span=12).mean())
    df['EMA_26'] = df.groupby('ticker')['Close'].transform(lambda x: x.ewm(span=26).mean())
    df['MACD'] = df['EMA_12'] - df['EMA_26']
    df['MACD_Signal'] = df.groupby('ticker')['MACD'].transform(lambda x: x.ewm(span=9).mean())
    df['MACD_Diff'] = df['MACD'] - df['MACD_Signal']
    
    # Volume
    df['Volume_MA'] = df.groupby('ticker')['Volume'].transform(lambda x: x.rolling(20, min_periods=20).mean())
    df['Volume_Ratio'] = df['Volume'] / (df['Volume_MA'] + 1e-10)
    
    # Volatilidade
    df['Volatility_5d'] = df.groupby('ticker')['Return_1d'].transform(lambda x: x.rolling(5, min_periods=5).std())
    df['Volatility_20d'] = df.groupby('ticker')['Return_1d'].transform(lambda x: x.rolling(20, min_periods=20).std())
    
    # Range
    df['Daily_Range'] = (df['High'] - df['Low']) / (df['Close'] + 1e-10)
    df['Daily_Range_MA'] = df.groupby('ticker')['Daily_Range'].transform(lambda x: x.rolling(10, min_periods=10).mean())
    
    # Decomposi√ß√£o Tend√™ncia + Ru√≠do
    df['Trend'] = df.groupby('ticker')['Close'].transform(lambda x: x.rolling(20, min_periods=20).mean())
    df['Noise'] = df['Close'] - df['Trend']
    df['Volatility'] = df.groupby('ticker')['Close'].transform(lambda x: x.pct_change().rolling(20, min_periods=20).std())
    
    return df

df_features = create_features_no_leakage(df_raw)

# TARGET: Retorno futuro (shift -1 = pr√≥ximo dia)
df_features['Future_Return'] = df_features.groupby('ticker')['Close'].pct_change().shift(-1)
df_features['Target_Binary'] = (df_features['Future_Return'] > 0).astype(int)

# Remover NaN e √∫ltimas linhas (sem target)
df_clean = df_features.dropna()
print(f"‚úì Features criadas: {df_features.shape[1]} colunas")
print(f"‚úì Dataset limpo: {len(df_clean)} amostras")
```

```python
# ============================================================================
# 3. AN√ÅLISE EXPLORAT√ìRIA
# ============================================================================
print("\n[3/10] AN√ÅLISE EXPLORAT√ìRIA")
print("-"*80)

print(f"Distribui√ß√£o do Target:")
print(df_clean['Target_Binary'].value_counts())
print(f"\nBalanceamento: {df_clean['Target_Binary'].mean():.2%} classe positiva")
```

```python
# ============================================================================
# 4. SPLIT TEMPORAL (80/20)
# ============================================================================
print("\n[4/10] SPLIT TEMPORAL")
print("-"*80)


df_clean = df_clean.sort_values('Date')
split_date = df_clean['Date'].quantile(0.8)

train_mask = df_clean['Date'] <= split_date
test_mask = df_clean['Date'] > split_date

feature_cols = [
    'Return_1d', 'Return_5d', 'Return_10d',
    'SMA_Cross_5_20', 'SMA_Cross_10_50',
    'RSI_14', 'BB_Position', 'MACD_Diff',
    'Volume_Ratio', 'Volatility_5d', 'Volatility_20d',
    'Daily_Range', 'Daily_Range_MA',
    'Trend', 'Noise', 'Volatility'
]

X_train = df_clean.loc[train_mask, feature_cols]
X_test = df_clean.loc[test_mask, feature_cols]
y_train = df_clean.loc[train_mask, 'Target_Binary']
y_test = df_clean.loc[test_mask, 'Target_Binary']

print(f"‚úì Treino: {len(X_train)} samples")
print(f"  {df_clean[train_mask]['Date'].min().date()} ‚Üí {df_clean[train_mask]['Date'].max().date()}")
print(f"‚úì Teste: {len(X_test)} samples")
print(f"  {df_clean[test_mask]['Date'].min().date()} ‚Üí {df_clean[test_mask]['Date'].max().date()}")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print("\n[5/10] BASELINES DE COMPARA√á√ÉO")
print("-"*80)

# Baseline 1: Prever sempre a classe majorit√°ria
baseline_majority = (y_test == y_test.mode()[0]).mean()
print(f"Baseline 1 (Classe Majorit√°ria): {baseline_majority:.4f}")

# Baseline 2: Momentum simples (se subiu ontem, vai subir hoje)
last_return = df_clean.loc[test_mask, 'Return_1d']
baseline_momentum = (y_test == (last_return > 0).astype(int)).mean()
print(f"Baseline 2 (Momentum Simples): {baseline_momentum:.4f}")

# Baseline 3: Random (50/50)
baseline_random = 0.5
print(f"Baseline 3 (Random): {baseline_random:.4f}")
```

```python
# ============================================================================
# 6. TREINAMENTO COM M√öLTIPLOS MODELOS
# ============================================================================
print("\n[6/10] TREINAMENTO DE MODELOS")
print("-"*80)

models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(n_estimators=100, max_depth=5, random_state=42, eval_metric='logloss'),
    'LightGBM': LGBMClassifier(n_estimators=100, random_state=42, verbose=-1)
}

results = {}
for name, model in models.items():
    print(f"\n{name}...")
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]
    
    results[name] = {
        'accuracy': accuracy_score(y_test, y_pred),
        'balanced_accuracy': balanced_accuracy_score(y_test, y_pred),
        'f1_score': f1_score(y_test, y_pred),
        'auc_roc': roc_auc_score(y_test, y_pred_proba),
        'model': model,
        'predictions': y_pred,
        'probabilities': y_pred_proba
    }
    
    print(f"  Acc: {results[name]['accuracy']:.4f} | "
          f"Bal.Acc: {results[name]['balanced_accuracy']:.4f} | "
          f"F1: {results[name]['f1_score']:.4f} | "
          f"AUC: {results[name]['auc_roc']:.4f}")
```

```python
# ============================================================================
# 7. TESTES ESTAT√çSTICOS DE SIGNIFIC√ÇNCIA
# ============================================================================
print("\n[7/10] TESTES ESTAT√çSTICOS DE SIGNIFIC√ÇNCIA")
print("-"*80)

best_model_name = max(results, key=lambda x: results[x]['balanced_accuracy'])
best_bal_acc = results[best_model_name]['balanced_accuracy']

print(f"\nMelhor modelo: {best_model_name}")
print(f"Balanced Accuracy: {best_bal_acc:.4f}")

# Teste 1: Permutation Test (o modelo √© melhor que acaso?)
print("\n--- Permutation Test ---")
def statistic(x, y, axis):
    return balanced_accuracy_score(x, y)

y_pred_best = results[best_model_name]['predictions']

# Criar fun√ß√£o para permutation test
def permutation_balanced_acc(y_true, y_pred_fixed, n_permutations=1000):
    """Testa se balanced accuracy √© significativamente melhor que acaso"""
    observed = balanced_accuracy_score(y_true, y_pred_fixed)
    
    permuted_scores = []
    for _ in range(n_permutations):
        y_permuted = np.random.permutation(y_true)
        score = balanced_accuracy_score(y_permuted, y_pred_fixed)
        permuted_scores.append(score)
    
    p_value = (np.sum(np.array(permuted_scores) >= observed) + 1) / (n_permutations + 1)
    return observed, p_value, permuted_scores

observed_score, p_value, perm_scores = permutation_balanced_acc(
    y_test.values, y_pred_best, n_permutations=1000
)

print(f"Balanced Accuracy Observada: {observed_score:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("‚úì Resultado SIGNIFICATIVO (p < 0.05)")
else:
    print("‚úó Resultado N√ÉO significativo (p ‚â• 0.05)")
    print("  ATEN√á√ÉO: Modelo pode n√£o ter skill preditivo real!")

# Teste 2: Bootstrap Confidence Intervals
print("\n--- Bootstrap Confidence Intervals (95%) ---")
bootstrap_scores = []
n_bootstrap = 1000

for i in range(n_bootstrap):
    indices = resample(range(len(y_test)), random_state=i)
    y_test_boot = y_test.iloc[indices]
    y_pred_boot = y_pred_best[indices]
    score = balanced_accuracy_score(y_test_boot, y_pred_boot)
    bootstrap_scores.append(score)

ci_lower = np.percentile(bootstrap_scores, 2.5)
ci_upper = np.percentile(bootstrap_scores, 97.5)

print(f"Balanced Accuracy: {best_bal_acc:.4f}")
print(f"95% CI: [{ci_lower:.4f}, {ci_upper:.4f}]")
print(f"Largura do intervalo: {ci_upper - ci_lower:.4f}")

if ci_lower > baseline_random:
    print(f"‚úì Intervalo n√£o inclui 0.5 (baseline random)")
else:
    print(f"‚úó ATEN√á√ÉO: Intervalo inclui 0.5 - modelo pode ser apenas acaso!")
```

```python
print("\n[8/10] WALK-FORWARD VALIDATION")
print("-"*80)

tscv = TimeSeriesSplit(n_splits=5)
wf_scores = []

best_model_class = type(results[best_model_name]['model'])
best_model_params = results[best_model_name]['model'].get_params()

print(f"Realizando 5 splits temporais com {best_model_name}...")

for fold, (train_idx, val_idx) in enumerate(tscv.split(X_train_scaled)):
    X_tr, X_val = X_train_scaled[train_idx], X_train_scaled[val_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]
    
    model_wf = best_model_class(**best_model_params)
    model_wf.fit(X_tr, y_tr)
    y_pred_wf = model_wf.predict(X_val)
    
    score = balanced_accuracy_score(y_val, y_pred_wf)
    wf_scores.append(score)
    print(f"  Fold {fold+1}: {score:.4f}")

print(f"\nM√©dia Walk-Forward: {np.mean(wf_scores):.4f} ¬± {np.std(wf_scores):.4f}")
print(f"Compara√ß√£o com teste √∫nico: {best_bal_acc:.4f}")

if abs(np.mean(wf_scores) - best_bal_acc) > 0.05:
    print("‚ö† ATEN√á√ÉO: Grande diferen√ßa entre WF e teste √∫nico - poss√≠vel overfitting!")
```

```python
print("\n[9/10] AN√ÅLISE CR√çTICA DOS RESULTADOS")
print("="*80)

print("\nüìä SUM√ÅRIO COMPARATIVO:")
print(f"{'M√©todo':<30} {'Balanced Accuracy':>18}")
print("-"*50)
print(f"{'Baseline Random (0.5)':<30} {baseline_random:>18.4f}")
print(f"{'Baseline Majorit√°ria':<30} {baseline_majority:>18.4f}")
print(f"{'Baseline Momentum':<30} {baseline_momentum:>18.4f}")
print(f"{best_model_name + ' (melhor ML)':<30} {best_bal_acc:>18.4f}")
print(f"{'Walk-Forward Validation':<30} {np.mean(wf_scores):>18.4f}")

print("\nüîç INTERPRETA√á√ÉO:")

improvement = (best_bal_acc - baseline_random) * 100
print(f"\n1. MAGNITUDE DO EFEITO:")
print(f"   Melhoria sobre baseline random: {improvement:.2f} pontos percentuais")

if improvement < 3:
    print("   ‚ö† Melhoria MUITO PEQUENA - efetivamente indistingu√≠vel do acaso")
elif improvement < 5:
    print("   ‚ö† Melhoria PEQUENA - skill preditivo limitado")
elif improvement < 10:
    print("   ‚úì Melhoria MODERADA - algum skill preditivo detect√°vel")
else:
    print("   ‚úì‚úì Melhoria SIGNIFICATIVA - skill preditivo robusto")

print(f"\n2. SIGNIFIC√ÇNCIA ESTAT√çSTICA:")
print(f"   P-value (permutation test): {p_value:.4f}")
if p_value < 0.01:
    print("   ‚úì‚úì Altamente significativo (p < 0.01)")
elif p_value < 0.05:
    print("   ‚úì Significativo (p < 0.05)")
else:
    print("   ‚úó N√ÉO significativo - resultados podem ser devidos ao acaso")

print(f"\n3. ESTABILIDADE:")
wf_std = np.std(wf_scores)
if wf_std < 0.02:
    print(f"   ‚úì Baixa variabilidade (std={wf_std:.4f}) - modelo est√°vel")
elif wf_std < 0.05:
    print(f"   ‚ö† Variabilidade moderada (std={wf_std:.4f})")
else:
    print(f"   ‚úó Alta variabilidade (std={wf_std:.4f}) - resultados inst√°veis")

print("\n4. IMPLICA√á√ïES PR√ÅTICAS:")
if best_bal_acc < 0.55 or p_value >= 0.05:
    print("   ‚ö† CONCLUS√ÉO: Os resultados sugerem que:")
    print("   ‚Ä¢ As features t√©cnicas capturaram sinais muito fracos")
    print("   ‚Ä¢ O ru√≠do domina a componente de tend√™ncia")
    print("   ‚Ä¢ Mercados podem estar pr√≥ximos da efici√™ncia informacional")
    print("   ‚Ä¢ Este √© um resultado ESPERADO e academicamente v√°lido")
    print("   ‚Ä¢ Demonstra os limites da previsibilidade em mercados financeiros")
else:
    print("   ‚úì Modelo demonstra algum skill preditivo, mas:")
    print("   ‚Ä¢ Ainda longe de ser utiliz√°vel para trading real")
    print("   ‚Ä¢ Custos de transa√ß√£o n√£o foram considerados")
    print("   ‚Ä¢ Resultados em per√≠odo espec√≠fico (2021-2026)")
```
